```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\joshd\Downloads\Religion by constituency.csv")
```

```python
df.head(10)
```

```python
df.rename(columns={'Religion (10 categories)':'Religion', 'Religion (10 categories) Code':'ReligionCode','Post-2019 Westminster Parliamentary constituencies':'ConstituencyName','Post-2019 Westminster Parliamentary constituencies Code':'ConstituencyCode' },inplace=True)
```

```python
df.head(5)
```

```python
total=df.groupby("ConstituencyCode").Observation.sum()
total.head(5)
```

```python
muslim=df.loc[df.Religion=='Muslim'].iloc[:,[0,4]]
muslim.head(5)
```

```python
percentage=(muslim.Observation/total.values)*100
percentagebycon=pd.concat([percentage,muslim.ConstituencyCode,df.ConstituencyName],axis=1)
percentagebycon=percentagebycon.iloc[0:575]
percentagebycon.head(5)
```

```python
percentagebycon.to_csv(r"C:\Users\joshd\Downloads\Muslim%_Constituency.csv",index=False)
```
